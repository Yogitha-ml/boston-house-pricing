## Independent and dependant features

```python
x= dataset.iloc[:,:-1]
y= dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size= 0.3,random_state= 42)
```

```python
x_train
```

### Standardize the data set**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regg = LinearRegression()
```

```python
regg.fit(x_train,y_train)
```

```python
## Print coefficients and intercept
```

```python
print(regg.coef_)
```

```python
print(regg.intercept_)
```

```python
regg.get_params()
```

```python
## prediction with test data
reg_pred=regg.predict(x_test)
```

```python
reg_pred
```

```python
## plot a scater plot fore pred
plt.scatter(y_test,reg_pred)
```

```python
## residuals
residuals=y_test-reg_pred
```

```python
residuals
```

```python
## Plot this residuals
sns.displot(residuals,kind='kde')
```

```python
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

## R Square and Adjusted R Sq

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
1-(1-score)*(len(y_test)-1/(len(y_test)-x_test.shape[1]-1))
```

## LInear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
mse=cross_val_score(lin_reg,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

## Ridge regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={'alpha':[12-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
ridge_reg = GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_reg.fit(x_train,y_train)
```

```python
print(ridge_reg.best_params_)
print(ridge_reg.best_score_)
```

## Lasso Regresion

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
params={'alpha':[12-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
lasso_reg = GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_reg.fit(x_train,y_train)
```

```python
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```

```python
y_pred=lasso_reg.predict(x_test)
from sklearn.metrics import r2_score
r2_score1=r2_score(y_pred,y_test)
```

```python
print(r2_score1)
```
